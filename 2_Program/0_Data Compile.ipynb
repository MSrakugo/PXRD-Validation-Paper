```python
import os
import shutil
import glob
import pandas as pd
import matplotlib.pyplot as plt
import re

import numpy as np
import math
```

```python
src_dir=""
# Compile
dst_dir = os.path.join(src_dir, "Compile")

# Compile
if not os.path.exists(dst_dir):
    os.makedirs(dst_dir)
    print(f'Created folder: {dst_dir}')
    
# 
for file_path in glob.glob("*/*/*.sum"):
    # 
    dst_file_path = os.path.join(dst_dir, os.path.basename(file_path))
    print(dst_file_path)
    
    # folder
    if "4Y_1" in os.path.basename(file_path):
        path_name_split = dst_file_path.split("/")
        
        file_num = path_name_split[-1].split("-")[-1]
        file_name = "20240219-3B_1-"+file_num
        dst_file_path=os.path.join(dst_dir, file_name)    

    # 
    shutil.copy2(file_path, dst_file_path)
    print(f'Copied: {file_path} to {dst_file_path}')
```

```python
XRD_pattern = pd.DataFrame()
for file_name in glob.glob("Compile/*.sum"):
    # 
    if (file_name != "Compile/20240219-2G-49.sum") and (file_name !="Compile/20240219-4B-1-10.sum"):
        data = pd.read_table(file_name, sep="	", header=None, index_col=0, names=["Intensity"])
        print(file_name)
        
        # file name
        file_name_title = file_name.split("/")[-1].split(".")[-2]
        print(file_name_title)
        
        # DataFrameDataFrame
        new_XRD_pattern = XRD_pattern.copy()
        # 
        new_XRD_pattern[file_name_title] = data
        # DataFrame
        XRD_pattern = new_XRD_pattern
```

```python
# Compile
dst_dir = os.path.join(src_dir, "Compile_Transformed_Data")

# Compile
if not os.path.exists(dst_dir):
    os.makedirs(dst_dir)
    print(f'Created folder: {dst_dir}')
    

def extract_number(sample): # index
    match = re.search(r'\d+$', sample)
    if match:
        return int(match.group())
    else:
        return float('inf')  #     
    
# index
sorted_samples = sorted(XRD_pattern.columns, key=extract_number)
print(sorted_samples)

XRD_pattern = XRD_pattern[sorted_samples]
XRD_pattern.to_excel("Compile_Transformed_Data/XRD_Data_ALL.xlsx") # SAVE
```

```python
######################################################
# Excel
file_path = '20240219_polaris_v1.5_.xlsm'
# Excel
xl = pd.ExcelFile(file_path)
# 
pattern = r'\d{4}-\d{1,2}-\d{1,2} \d{1,2}\.\d{2}\.\d{2}'
# 
matching_sheets = [sheet for sheet in xl.sheet_names if re.match(pattern, sheet)]


#########################################
need_col = ['', '', ''] # 
# 
data_frames = pd.DataFrame(columns=need_col)
for sheet in matching_sheets:
    #
    data_sheet = pd.read_excel(file_path, sheet_name=sheet, header=9, index_col=3)[need_col].dropna()
    data_frames = pd.concat([data_sheet, data_frames, ]) # compile file
#########################################
    
#########################################sample name
"""
MTNSP-MTN-??MTN-??

"""

sample_name_list = []
data_frames["file"]=np.nan
data_frames["sample"]=np.nan
########################################################filesample
for file_name in XRD_pattern.columns:
    
    #####################################################################file name 
    if "3B_1" in file_name: # file4Y_1
        file_name = '20240219-4Y_1-'+ file_name.split("-")[-1]
        #print(file_name)
    #####################################################################file name 
    
    # read sample name
    sample_name = data_frames.loc[file_name]['']
    
    #####################################################################sample name 
    if sample_name.split("-")[0] == "MTN": # "SP-"MTN
        sample_name = "SP-"+sample_name
        print(sample_name)
    if "SP8-MTN" in sample_name: # "SP8-MTN"
        sample_name = re.sub(r'^SP8', 'SP', sample_name)
        print(sample_name)
    if "SP8-HH" in sample_name: # "SP8-HH"
        sample_name = re.sub(r'^SP8', 'SP', sample_name)
        print(sample_name)
    if "SP-8-OU" in sample_name: # "SP-8-OU"
        sample_name = re.sub(r'^SP-8-OU', 'SP8-OU', sample_name)
        print(sample_name)
    if "SP-8_OU" in sample_name: # "SP-8-OU"
        sample_name = re.sub(r'^SP-8_OU', 'SP8-OU', sample_name)
        print(sample_name)
    #####################################################################sample name 

    sample_name_list.append(sample_name) # file namesample name
    data_frames.loc[file_name, "file"] = file_name
    data_frames.loc[file_name, "sample"] = sample_name
    data_frames.loc[file_name, "SP-NO"] = sample_name
########################################################filesample
    
XRD_pattern_rename = XRD_pattern.copy()
XRD_pattern_rename.columns=sample_name_list #sample name

########## index
XRD_pattern = XRD_pattern_rename[sample_name_list]
########## index

XRD_pattern_rename.to_excel("Compile_Transformed_Data/XRD_Data_ALL_Sample_name.xlsx") #
data_frames.to_excel("Compile_Transformed_Data/Sample_file_name_.xlsx") #
```

```python
# XRD_pattern_renamecolumns
sample_series = ["SP-MTN", "SP8-OU", "Okamoto_IODP", "SP-H-", "SP-Well-20", "SP-8-Otogo", "SP-HH-", "SP8-SY", "EH"]
# YoshidaK, Uno

sample_name_list_check = sample_name_list.copy()

n=0
for sample_series_name in sample_series:
    #    
    sample_series_list = [sample_name for sample_name in sample_name_list if sample_series_name in sample_name]
    
    ####################################### check sample: 
    for sample_series_list_name in sample_series_list:
        sample_name_list_check.remove(sample_series_list_name)
    print(sample_series_name)
    print(len(sample_series_list ))
    n=len(sample_series_list )+n
    ####################################### check sample: 

    ####################################### 
    sorted_samples = sorted(sample_series_list, key=extract_number)
    XRD_pattern_rename[sorted_samples].to_excel("Compile_Transformed_Data/Series_XRD/XRD_"+sample_series_name+".xlsx")
    
# check sample: CeO2STDCeO2OK
print(n)
print(sample_name_list_check)
```

```python
################################################ Diff Angle change
diffraction_angle_lambda1 = XRD_pattern.index

lambda1 = 0.6999200 # refinement result by manually 0.6999200 ver 240304
lambda2 = 1.5418

diffraction_angle_lambda1 = diffraction_angle_lambda1/2 # 2theta  theta
diffraction_angle_lambda1_radians = [math.radians(angle) for angle in diffraction_angle_lambda1] # 
theta2_list = []

# 
for angle in diffraction_angle_lambda1_radians:
    # lambda2 * sin(angle) / lambda1  -1  1 
    sin_value = (lambda2 * math.sin(angle)) / lambda1
    if -1 <= sin_value <= 1:  # math.asin  [-1, 1]
        theta2 = math.degrees(math.asin(sin_value)) 
        theta2 = theta2*2 #theta  2theta
        theta2_list.append(theta2)
    else:
        # 
        pass
################################################ Diff Angle change

XRD_pattern_CuKa = XRD_pattern_rename.iloc[0:len(theta2_list)]
XRD_pattern_CuKa.index = theta2_list
XRD_pattern_CuKa.to_excel("Compile_Transformed_Data/XRD_Data_ALL_CuKa.xlsx") # SAVE
```

```python
for sample_name in XRD_pattern_rename.columns:
        print(sample_name)
        data = XRD_pattern_rename[sample_name].dropna()
        ######################################################## XRD pattern
        # XRD
        plt.figure(figsize=(30, 10))

        ############################ Plot experiment pattern
        plt.plot(data.index.values, data.values, label=sample_name, linewidth=0.5)
        ############################ Plot experiment pattern

        # 
        plt.xlabel(r'2$\theta$ (degrees)', fontsize=40)
        plt.ylabel('Intensity (a.u.)', fontsize=40)

        # xy
        plt.xlim([2, 40])
        try:
            plt.ylim([0, data.max(skipna=True)*1.05])
        except:
            pass

        # 
        plt.title(sample_name, fontsize=40)
        #plt.legend(False)#plt.legend(fontsize=40)

        # ticks
        plt.xticks(fontsize=25)
        plt.yticks(fontsize=25)

        plt.legend()
        plt.savefig("Compile_Transformed_Data/XRD_Pattern/"+sample_name+".pdf", bbox_inches="tight")
        plt.show()
        ######################################################## XRD pattern
```
